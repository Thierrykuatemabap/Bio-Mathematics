```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definition of function
def f(x, r):
    return r * x * (1 - x)

# Domain
x = np.linspace(0, 1, 400)

# Parameter of cobweb
x0 = 0.3
N = 5
r = 0.3

# Tracé des courbes
plt.figure(figsize=(8, 6))
plt.plot(x, f(x, r), label=r"$x_{n} = r * x * (1-x)$")
plt.plot(x, x, label=r"$y = x$")

# Cobwebbing
xn = x0
for i in range(N):
    yn = f(xn, r)
    plt.plot([xn, xn], [xn, yn])   # vertical
    plt.plot([xn, yn], [yn, yn])   # horizontal
    xn = yn

# Mise en forme
plt.xlabel("x")
plt.ylabel("y")
plt.title(r"Cobwebbing method for $x_{n} = r*x_{n-1} * (1-x_{n-1})$")
plt.legend()
plt.grid(True)
plt.show()

```

```python
R = [0.3, 2, 3.2, 4]
plt.figure(figsize=(10, 8))
for i, r in enumerate(R):
    plt.subplot(2, 2, i+1)
    
    plt.plot(x, f(x, r), label=r"$x_{n} = r * x * (1-x)$")
    plt.plot(x, x, label=r"$y = x$")

    # Cobwebbing
    xn = x0
    for i in range(N):
        yn = f(xn, r)
        plt.plot([xn, xn], [xn, yn])   # vertical
        plt.plot([xn, yn], [yn, yn])   # horizontal
        xn = yn

    # Mise en forme
    plt.xlabel("t")
    plt.ylabel("x")
    plt.title(f"behavior for r = {r}")
    plt.grid()
    plt.legend()
    
plt.tight_layout()
plt.show()
```

```python
# Burfication

rmin = 0.0
rmax = 3
n = 850
N = 100

R = np.zeros(n*N + 1)
X = np.zeros(n*N + 1)

plt.figure(figsize=(6,6))

for m in range(1, n+1):
    r = rmin + (rmax - rmin) * (m/n)
    x = 0.5

    for k in range(1, N+1):
        x = r * x * (1-x)
        X[k] = x
        R[k] = r

    plt.plot(R[0:N+1], X[0:N+1], '.', markersize=1, color='blue')

plt.xlabel("r")
plt.ylabel("x")
plt.title("behavior")
plt.xlim(rmin, rmax)
plt.ylim(0,1)
plt.grid()
plt.legend()
    
plt.tight_layout()
plt.show()
```
