```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Exo 1)

# Definition of function
def f(x):
    return x + x * (1 - x)

# Domain
x = np.linspace(0, 2.5, 400)

# Parameter of cobweb
x0 = 0.5
N = 20

# Tracé des courbes
plt.figure(figsize=(6, 6))
plt.plot(x, f(x), label=r"$y = x + x(1-x)$")
plt.plot(x, x, label=r"$y = x$")

# Cobwebbing
xn = x0
for i in range(N):
    yn = f(xn)
    plt.plot([xn, xn], [xn, yn])   # vertical
    plt.plot([xn, yn], [yn, yn])   # horizontal
    xn = yn

# Mise en forme
plt.xlabel("x")
plt.ylabel("y")
plt.title(r"Cobwebbing method for $x_{n+1} = x_n + x_n(1-x_n)$")
plt.legend()
plt.grid(True)
plt.show()

```

```python
import matplotlib.pyplot as plt

# Definition 
def f(x):
    if x > 0.25:
        return 0.5
    else:
        return 2 * x

# Paramètres
N = 15  # nomber of iteration 
x0_values = [0.05, 0.1, 0.2, 0.3, 0.6, 0.9]

plt.figure(figsize=(7, 5))

for x0 in x0_values:
    x = [x0]
    for n in range(N):
        x.append(f(x[-1]))
    plt.plot(range(len(x)), x, marker='o', label=f"$x_0 = {x0}$")

plt.xlabel("n")
plt.ylabel("$x_n$")
plt.title("Solutions of the differential equation")
plt.legend()
plt.grid(True)
plt.show()
```
